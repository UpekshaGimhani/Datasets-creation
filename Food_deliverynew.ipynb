```python
import pandas as pd

# Load the extended dataset (make sure this path matches your file location)
file_path = "Food_Delivery_Extended.csv"  # Make sure this file exists in the correct location
df = pd.read_csv(file_path)

# Some real-world restaurant chain names
real_restaurants = [
    "McDonald's", "KFC", "Burger King", "Pizza Hut", "Subway", "Domino's Pizza",
    "Starbucks", "Dunkin' Donuts", "Taco Bell", "Chipotle", "Chick-fil-A",
    "Wendy's", "Popeyes", "Five Guys", "Nando's", "Costa Coffee", "Panera Bread",
    "Shake Shack", "In-N-Out Burger", "Hardee's", "Arby's", "Little Caesars",
    "Wingstop", "Tim Hortons", "Jollibee", "Pret A Manger", "Buffalo Wild Wings",
    "Cold Stone Creamery", "Benihana", "Qdoba Mexican Eats", "Zaxby's", "El Pollo Loco"
]

# Assign real restaurant names cyclically
df["restaurant_name"] = [real_restaurants[i % len(real_restaurants)] for i in range(len(df))]

# Save the updated dataset
output_path = "Food_Delivery_Extended_RealNames.csv"
df.to_csv(output_path, index=False)

print(f"✅ Updated file saved as: {output_path}")
```
